```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wise = pd.read_csv('./wise.matrix')
wise_X = wise.drop(columns = ['Sample', 'Tumor type', 'Stage'])
display(wise_X.head())
```

```python
sns.set_theme(color_codes=True, style="white", context="talk", font_scale=0.8)
sns.set_palette(["#377eb8", "#e41a1c"], n_colors=2)
```

```python
important_columns = [
    "Cancer Status",
    "8:130000001-140000000",
    "8:140000001-150000000",
    "8:110000001-120000000",
    "8:90000001-100000000",
    "8:120000001-130000000",
    "8:70000001-80000000",
    "1:210000001-220000000",
    "8:20000001-30000000",
]
g = sns.PairGrid(wise_X[important_columns], hue="Cancer Status")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot, alpha=0.2, linewidth=0, s=30)
g.map_lower(sns.kdeplot, alpha=0.5)
plt.legend(loc="lower center", bbox_to_anchor=(-4, -0.7), ncol=2, labels=["Control", "Cancer"], frameon=False)
```

```python
g1 = sns.PairGrid(wise_X[important_columns[:6]], hue="Cancer Status")
g1.map_diag(sns.histplot)
g1.map_upper(sns.scatterplot, alpha=0.2, linewidth=0, s=20)
g1.map_lower(sns.kdeplot, alpha=0.5)
plt.legend(loc="lower center", bbox_to_anchor=(-2, -0.7), ncol=2, labels=["Control", "Cancer"], frameon=False)

plt.show()
```
